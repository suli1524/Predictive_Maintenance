```python
# create X and y features:

X = df.drop(columns=['Product ID', 'Machine failure'])
y = df['Machine failure']
```

```python
# run a train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y, train_size= .8, random_state=42)
```

#### using SMOTE to Balance Data

```python
# import libaries
from sklearn.datasets import make_classification
from collections import Counter
from imblearn.over_sampling import SMOTE
```

```python
# transform the dataset
oversample = SMOTE()
X, y = oversample.fit_resample(X, y)
```

```python
X.shape
```

```python
y.value_counts()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, train_size=.8, random_state = 42)
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
logreg.score(X_train,y_train), logreg.score(X_test,y_test)
```

#### Randomforest Model

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
train_predictions = rf.predict(X_train)
test_predictions = rf.predict(X_test)
print(f'train pred: {train_predictions:}')
print(f'test pred: {test_predictions}')
```

```python
train_accuracy = accuracy_score(y_train, train_predictions)
test_accuracy = accuracy_score(y_test, test_predictions)

print(f"Training Accuracy: {train_accuracy:.4f}")
print(f"Testing Accuracy: {test_accuracy:.4f}")
```

#### Random forest with XGboost

```python
# Create RandomForest model
rf_model = RandomForestClassifier(random_state=42)
# Create XGBoost model
xgb_model = XGBClassifier(random_state=42)
```

```python
# Define the parameter grid for RandomizedSearchCV
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [3, 5, 7],
    'max_features': ['auto', 'sqrt', 'log2'],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False],
    'class_weight': [None, 'balanced']
}
```

```python
# Create RandomizedSearchCV for RandomForest
rf_random_search = RandomizedSearchCV(
    estimator=rf_model,
    param_distributions=param_grid,
    n_iter=10,  # Adjust the number of iterations based on your computational resources
    scoring='accuracy',
    cv=5,  # Adjust the number of cross-validation folds as needed
    random_state=42,
    n_jobs=2  # Use all available cores
)
```

```python
# Fit RandomizedSearchCV for RandomForest
rf_random_search.fit(X_train, y_train)
```

```python
# Get the best parameters for RandomForest
best_rf_params = rf_random_search.best_params_
```

```python
best_rf_params
```

```python
# Create RandomForest model with the best parameters
best_rf_model = RandomForestClassifier(**best_rf_params, random_state=42)
```

```python
# Train the best RandomForest model
best_rf_model.fit(X_train, y_train)
```

```python
# Make predictions on the testing set
rf_predictions = best_rf_model.predict(X_test)
```

```python
# Evaluate RandomForest model
rf_accuracy = accuracy_score(y_test, rf_predictions)
print(f"RandomForest Model Accuracy: {rf_accuracy:.4f}")
```
