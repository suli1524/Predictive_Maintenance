### EDA: Exploratory Data Analysis

```python
# Creating a correlation heat map
correlation_matrix = df.corr()
machine_failure = correlation_matrix['Machine failure']
plt.figure(figsize=(10,8))
sns.heatmap(machine_failure.to_frame(), annot=True, cmap='coolwarm', cbar=True, linewidths=0.5)
plt.title("Correlation to machine failure")
plt.show()
```

```python
####create histogram for all numerical columns

num_columns = df.select_dtypes(include=['int64','float64'])
                                              
for column in num_columns.columns:
    plt.figure(figsize=(5,5))
    plt.hist(num_columns[column], bins=40,color='blue')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('frequency')
```

```python
sns.pairplot(df)
```

```python
# Distribution of machine failures
plt.figure(figsize=(8, 5))
sns.countplot(x='Machine failure', data=df)
plt.title('Distribution of Machine Failures')
plt.xlabel('Failure (1: Yes, 0: No)')
plt.ylabel('Count')
plt.show()
```
